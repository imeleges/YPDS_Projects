### Обработка пропущенных значений. 

Попробуем заменить пропущенные значения там, где это возможно и уместно. 
Если пропуски не имеют большой ценности или веса чтобы повлиять на результат исследования, такие данные можно отбросить  

Создадим удобную функцию для проверки количества пропусков и подсчета в процентном соотношении

```python
def cheking_isna_sum_and_percent_for(df_series):
    df_sum = df_series.isna().sum()
    df_percent = df_series.isna().mean()
    df_percent = "{:.2%}".format(df_percent)
    print(f"Количество пропущенных записей: {df_sum}\nЭто {df_percent} от общего числа {data.shape[0]}")
```

Видим что населенные пункты имеют буквы `е` и `ё` в словах `поселок` и `посёлок`, а так же есть различия в написании, например `Кудрово` и `деревня Кудрово`.  
Но все названия самих населенных пунктов начинаются с заглавной буквы, можем обработать функцией `drop_lower_cased` которая найдет все слова начинающиеся с маленькой буквы и отбросит их, вернув только слово с заглавной.  
Конечно, такие населенные пункты как `поселок городского типа имени Свердлова` потеряют все слова кроме самого `Свердлова`, но таких значений довольно низкий процент и кроме того само название сохранилось.

```python
def drop_lower_cased(str_cell):
    for i in range(len(str_cell)):
        if str_cell[i].isupper():
            return str_cell[i: len(str_cell)]
        
data['locality_name'] = data['locality_name'].apply(drop_lower_cased)
```

Самый нежелательный тип данных это `object`, но мы ничего с ним сделать не можем и хорошо что у нас имеется только одна такая колонка. Однако, мы можем обработать типы данных `float64` и `int64`.  
Напишем цикл, который получит названия каждой колонки ипроверит ее тип данных. С помощью конструкции `if else` проверим если тип данных равен `int64` или `float64`, и приведем к оптимальному типу данных согласно содержимому колонки методом `to_numeric` и параметром `downcast = 'integer'` и `downcast = 'float'` соответственно.

В случае если возникнет какая нибудь ошибка, вы выведем это на экран благодаря конструкции `try except`. 


```python
for column in data.columns:
    
    if data[column].dtypes == "int64":
        try:
            data[column] = pd.to_numeric(data[column], downcast = 'integer')
        except:
            print(f"Пропущена колонка {column} с типом {data[column].dtypes}")
    elif data[column].dtypes == "float64":
        try:
            data[column] = pd.to_numeric(data[column], downcast = 'float')
        except:
            print(f"Пропущена колонка {column} с типом {data[column].dtypes}")
```

### Этаж квартиры; варианты — первый, последний, другой

Чтобы узнать типы этажей, то для последнего этажа достаточно сравнить если он равен общему количеству этажей, а первый этаж если равен 1. Все остальное попадет в категорию "Другой"

```python
data[['floors_total', 'floor']].sample(5, random_state = 5)
```

Создадим функцию <code>floortype_finder</code> которая будет принимать датафрейм и отсортировывать данные из колонок и с помощью цикла  возвращать результат сравнения в виде строки.

```python
def floortype_finder(data):
    
    floors_total = data['floors_total']
    floor = data['floor']
    
    if floor == 1:
        return "Первый"
    elif floor == floors_total:
        return "Последний"
    return "Другой"
```

Добавим новую колонку <code>data['floor_type']</code> и запишем результат выполнения функции с помощью <code>.apply()</code> где в качестве аргумента укажем <code>axis=1</code> для построчного исполнения.  

Выведем 10 случайных записей по этим трем колонкам 

```python
data['floor_type'] = data.apply(floortype_finder, axis=1)
data[['floors_total', 'floor', 'floor_type']].sample(10)
```

Напишем кастомизированную функцию для подсчета максимального и минимального значения усов для диаграммы размаха. Подсчитаем количество элементов после максимального значения и вычислим процент от общего числа записей

```python
def whiskers_limits(series):
    q1 = np.percentile(series, 25)  # Q1
    q2 = np.percentile(series, 50)  # Q2 (Median)
    q3 = np.percentile(series, 75)  # Q3
    whisker_maximum = round(q3 + ((q3 - q1) * 1.5), 2) # “maximum”: Q3 + 1.5 * IQR 
    whisker_minimum = round(q1 - ((q3 - q1) / 1.5), 2) # “minimum”: Q1 - 1.5 * IQR
    above_whisker_maximum = data.query('@series > @whisker_maximum')[series.name].count()
    total_amount = data[series.name].count()
    
    print(f"{whisker_minimum} - Минимальное значение для левого 'уса'\n"
          f"{whisker_maximum} - Максимальное значение для правого 'уса',\n"
          f"{above_whisker_maximum} (или {round(above_whisker_maximum / total_amount * 100, 2)}%)"
          f" из {total_amount} - N элементов после правого 'уса'")
```

#### Площадь квартир в центре

Для центральной области модифицируем функцию <code>whiskers_limits_central</code> для определения границ усов 

```python
def whiskers_limits_central(series):
    q1 = np.percentile(series, 25)  # Q1
    q2 = np.percentile(series, 50)  # Q2 (Median)
    q3 = np.percentile(series, 75)  # Q3
    whisker_maximum = round(q3 + ((q3 - q1) * 1.5), 2) # “maximum”: Q3 + 1.5 * IQR 
    whisker_minimum = round(q1 - ((q3 - q1) / 1.5), 2) # “minimum”: Q1 - 1.5 * IQR
    above_whisker_maximum = spb_central.query('@series > @whisker_maximum')[series.name].count()
    total_amount = spb_central[series.name].count()
    
    print(f"{whisker_minimum} - Минимальное значение для левого 'уса'\n"
          f"{whisker_maximum} - Максимальное значение для правого 'уса',\n"
          f"{above_whisker_maximum} (или {round(above_whisker_maximum / total_amount * 100, 2)}%)"
          f" из {total_amount} - N элементов после правого 'уса'")
```

Построим и выведем на экран гистограмму по общей площади, изучим какие и как часто встречаются значения.
