## Описание этапа обработки

- Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. 
- Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

- Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы.
- Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

- Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. 
- Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом.
- Затем его перемешивают и снова проводят измерения. 
Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. *slab*, «плита»).

Переименуем колонки

```python
data_arc.columns = ["key", "start_heating", "end_heating", "active_power", "reactive_power"]
```

Обработаем аномальное значение, посмотрим сколько записей есть с отрицательным показателем в колонке `reactive_power`

```python
data_arc[data_arc["reactive_power"] <= 0]
```

```python
data_arc[data_arc["key"] == 2116]
```

Имеются 4 записи с `key = 2116`, но удалим только одну запись с выбросом индекс которой равен `9780`

```python
data_arc = data_arc.drop(9780)
```

Добавим новую колонку `full_power` с полной мощностью, которая равна корню из суммы квадратов `active_power` и `reactive_power` мощностей

```python
data_arc["full_power"] = np.sqrt(data_arc["active_power"]**2 + data_arc["reactive_power"]**2)
```

Так же добавим колонку `heating_duration_sec` с продолжительностью нагрева в секундах

```python
data_arc["heating_duration_sec"] = (
    (data_arc["end_heating"] - data_arc["start_heating"])
     .dt.total_seconds().astype('int64')
)
```

Выведем первые 10 записей обновленного датафрейма и построим график распределения полной мощности

```python
data_arc.head(10)
```

Для общей таблицы нам потребуются числовые значения, объединенные по `key` который имеет несколько записей с разными, промежуточными данными. По этому добавим еще несколько колонок, суммы колонок `heating_duration_sec` и `full_power`

```python
data_arc["total_heating_duration"] = data_arc.groupby(data_arc["key"])["heating_duration_sec"].transform("sum")
data_arc["total_full_power"] = data_arc.groupby(data_arc["key"])["full_power"].transform("sum")

data_arc.head(10)
```

Подготовим эти данные, сгруппируем и оставим только колонки суммарной полной мощности и времени нагрева.

```python
data_arc_total = data_arc[["key", "total_full_power", "total_heating_duration"]].groupby("key").nth(0).reset_index()
data_arc_total.head()
```

По такому же принципу возьмем начальные и конечные показатели для активной и реактивной можности, сгруппировав в одтельный датафрейм.

```python
data_arc_start = data_arc.groupby("key")[["active_power", "reactive_power"]].nth(0)
data_arc_finish = data_arc.groupby("key")[["active_power", "reactive_power"]].nth(-1)
data_arc_power = data_arc_start.merge(data_arc_finish, on="key", how="left").reset_index()
```

```python
data_arc_power.head()
```

Теперь соединим их вместе и переименуем колонки

```python
data_arc_grouped = data_arc_power.merge(data_arc_total, on="key", how="left")
data_arc_grouped.columns = ["key",
                            "start_active_power",
                            "start_reactive_power",
                            "end_active_power",
                            "end_reactive_power",
                            "total_full_power",
                            "total_heating_duration"]
```

И так, новый датафрейм готов, его будем испольовать в объединенном датафрейме перед началом построения моделей.

```python
data_arc_grouped.head()
```

### Подбор и обучение моделей

Возьмем три модели `LinearRegression`, `RandomForestRegressor` и `CatBoostRegressor`  и подберем гиперпараметры на обущающей выборке `features_train`, `target_train`

#### LinearRegression

```python
model_LR = LinearRegression()
model_LR.fit(features_train, target_train)
```

#### RandomForestRegressor

```python
gridParams = {"max_depth" : range(14, 16),
              "n_estimators" : range(259, 261)}

model_forest = RandomForestRegressor(random_state=12345)
grid = GridSearchCV(model_forest, gridParams, cv=10, scoring="neg_mean_squared_error")
grid.fit(features_train, target_train)

print(grid.best_params_)
```

#### CatBoostRegressor

```python
model_CatBoost = CatBoostRegressor(loss_function="MAE",
                                   random_state=12345,
                                   logging_level="Silent")

grid = {"n_estimators": [1000], "learning_rate": [0.1, 0.2, 0.03], "depth": [2, 10]}

grid_search_result = model_CatBoost.grid_search(grid, 
                                       X=features_train, 
                                       y=target_train, verbose=False)

print(grid_search_result["params"])
```

### Тестирование моделей

Модели обучены, теперь проверим результат `MAE` на тестовой выборке и выберем лучшую модель с самым низки показателем.

#### LinearRegression

```python
predictions_test = model_LR.predict(features_test)

result = mean_absolute_error(target_test, predictions_test)
print(f"Значение MAE: {result}")
```

#### RandomForestRegressor

```python
model_forest = RandomForestRegressor(max_depth=14, n_estimators=259, random_state=12345)
model_forest.fit(features_train, target_train)

predictions_test = model_forest.predict(features_test)

result = mean_absolute_error(target_test, predictions_test)
print(f"Значение MAE: {result}")
```

#### CatBoostRegressor

```python
model_CatBoost = CatBoostRegressor(loss_function="RMSE",
                                   random_state=12345,
                                   learning_rate=0.03,
                                   depth=4,
                                   iterations=1000)

model_CatBoost.fit(features_train, target_train, verbose=100)
predict_test = model_CatBoost.predict(features_test)
print(f"Значение MAE: {mean_absolute_error(target_test, predict_test)}")
```

Проверив все три модели, можно выбрать `CatBoostRegressor` с `MAE = 5.64`
