```python
data_arc["heating_duration_sec"] = (
    (data_arc["end_heating"] - data_arc["start_heating"])
     .dt.total_seconds().astype('int64')
)
```

–í—ã–≤–µ–¥–µ–º –ø–µ—Ä–≤—ã–µ 10 –∑–∞–ø–∏—Å–µ–π –æ–±–Ω–æ–≤–ª–µ–Ω–Ω–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ–ª–Ω–æ–π –º–æ—â–Ω–æ—Å—Ç–∏

```python
data_arc.head(10)
```

–î–ª—è –æ–±—â–µ–π —Ç–∞–±–ª–∏—Ü—ã –Ω–∞–º –ø–æ—Ç—Ä–µ–±—É—é—Ç—Å—è —á–∏—Å–ª–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è, –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω—ã–µ –ø–æ `key` –∫–æ—Ç–æ—Ä—ã–π –∏–º–µ–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ –∑–∞–ø–∏—Å–µ–π —Å —Ä–∞–∑–Ω—ã–º–∏, –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏. –ü–æ —ç—Ç–æ–º—É –¥–æ–±–∞–≤–∏–º –µ—â–µ –Ω–µ—Å–∫–æ–ª—å–∫–æ –∫–æ–ª–æ–Ω–æ–∫, —Å—É–º–º—ã –∫–æ–ª–æ–Ω–æ–∫ `heating_duration_sec` –∏ `full_power`

```python
data_arc["total_heating_duration"] = data_arc.groupby(data_arc["key"])["heating_duration_sec"].transform("sum")
data_arc["total_full_power"] = data_arc.groupby(data_arc["key"])["full_power"].transform("sum")

data_arc.head(10)
```

–ü–æ–¥–≥–æ—Ç–æ–≤–∏–º —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ, —Å–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –∏ –æ—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –∫–æ–ª–æ–Ω–∫–∏ —Å—É–º–º–∞—Ä–Ω–æ–π –ø–æ–ª–Ω–æ–π –º–æ—â–Ω–æ—Å—Ç–∏ –∏ –≤—Ä–µ–º–µ–Ω–∏ –Ω–∞–≥—Ä–µ–≤–∞.

```python
data_arc_total = data_arc[["key", "total_full_power", "total_heating_duration"]].groupby("key").nth(0).reset_index()
data_arc_total.head()
```

–ü–æ —Ç–∞–∫–æ–º—É –∂–µ –ø—Ä–∏–Ω—Ü–∏–ø—É –≤–æ–∑—å–º–µ–º –Ω–∞—á–∞–ª—å–Ω—ã–µ –∏ –∫–æ–Ω–µ—á–Ω—ã–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –¥–ª—è –∞–∫—Ç–∏–≤–Ω–æ–π –∏ —Ä–µ–∞–∫—Ç–∏–≤–Ω–æ–π –º–æ–∂–Ω–æ—Å—Ç–∏, —Å–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞–≤ –≤ –æ–¥—Ç–µ–ª—å–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º.

### –ü–æ–¥–±–æ—Ä –∏ –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

–í–æ–∑—å–º–µ–º —Ç—Ä–∏ –º–æ–¥–µ–ª–∏ `LinearRegression`, `RandomForestRegressor` –∏ `CatBoostRegressor`  –∏ –ø–æ–¥–±–µ—Ä–µ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –Ω–∞ –æ–±—É—â–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ `features_train`, `target_train`

#### LinearRegression

```python
model_LR = LinearRegression()
model_LR.fit(features_train, target_train)
```

#### RandomForestRegressor

```python
gridParams = {"max_depth" : range(14, 16),
              "n_estimators" : range(259, 261)}

model_forest = RandomForestRegressor(random_state=12345)
grid = GridSearchCV(model_forest, gridParams, cv=10, scoring="neg_mean_squared_error")
grid.fit(features_train, target_train)

print(grid.best_params_)
```

#### CatBoostRegressor

```python
model_CatBoost = CatBoostRegressor(loss_function="MAE",
                                   random_state=12345,
                                   logging_level="Silent")

grid = {"n_estimators": [1000], "learning_rate": [0.1, 0.2, 0.03], "depth": [2, 10]}

grid_search_result = model_CatBoost.grid_search(grid, 
                                       X=features_train, 
                                       y=target_train, verbose=False)

print(grid_search_result["params"])
```

### –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

–ú–æ–¥–µ–ª–∏ –æ–±—É—á–µ–Ω—ã, —Ç–µ–ø–µ—Ä—å –ø—Ä–æ–≤–µ—Ä–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç `MAE` –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ –∏ –≤—ã–±–µ—Ä–µ–º –ª—É—á—à—É—é –º–æ–¥–µ–ª—å —Å —Å–∞–º—ã–º –Ω–∏–∑–∫–∏ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–º.

#### LinearRegression

```python
predictions_test = model_LR.predict(features_test)

result = mean_absolute_error(target_test, predictions_test)
print(f"–ó–Ω–∞—á–µ–Ω–∏–µ MAE: {result}")
```

#### RandomForestRegressor

```python
model_forest = RandomForestRegressor(max_depth=14, n_estimators=259, random_state=12345)
model_forest.fit(features_train, target_train)

predictions_test = model_forest.predict(features_test)

result = mean_absolute_error(target_test, predictions_test)
print(f"–ó–Ω–∞—á–µ–Ω–∏–µ MAE: {result}")
```

#### CatBoostRegressor

```python
model_CatBoost = CatBoostRegressor(loss_function="RMSE",
                                   random_state=12345,
                                   learning_rate=0.03,
                                   depth=4,
                                   iterations=1000)

model_CatBoost.fit(features_train, target_train, verbose=100)
predict_test = model_CatBoost.predict(features_test)
print(f"–ó–Ω–∞—á–µ–Ω–∏–µ MAE: {mean_absolute_error(target_test, predict_test)}")
```

–ü—Ä–æ–≤–µ—Ä–∏–≤ –≤—Å–µ —Ç—Ä–∏ –º–æ–¥–µ–ª–∏, –º–æ–∂–Ω–æ –≤—ã–±—Ä–∞—Ç—å `CatBoostRegressor` —Å `MAE = 5.64`
