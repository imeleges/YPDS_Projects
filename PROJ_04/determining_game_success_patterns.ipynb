Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, *Xbox* или *PlayStation*).

### Обработка - `year_of_release`

Проверим сколько пропущеных значений в данной колонке

```python
df_games['year_of_release'].isna().sum()
```

Т.к. пропущенных данных довольно много, посмотрим на уникальные значения чтобы представить как обработать их дальше

```python
df_games['year_of_release'].unique()
```

Среди цифровых значений года выпуска игры есть пропуски обозначенные как `nan`, отсортируем их в отдельный датафрейм.

```python
df_missing_year = df_games.query('year_of_release == "NaN"')[['name', 'platform', 'year_of_release']]
```

Напишем функцию `year_finder`, которая получая имя игры проверит если такая же игра выходила на других платформах. Обычно, разница между годами не существенная примерно 1-3 года, по этому найдя игру на других платформах все же возмем медианое значение и заполним им пропуски для этой игры.

```python
def year_finder(gamename):
    median = df_games[df_games['name'] == gamename]['year_of_release'].median()
    df_games.loc[(df_games['name'] == gamename)
                 & (df_games['year_of_release'].isna()), 'year_of_release'] = np.round(median)
```

Переберем все уникальные имена игр из `df_missing_year` и применим функцию `year_finder`

```python
for gamename in df_missing_year['name'].unique():
    year_finder(gamename)
```

Узнаем если остались игры с пропусками

```python
df_games['year_of_release'].isna().sum()
```

### Оптимизация типов данных

После заполнения пропусков, посмотрим какие у нас есть типы данных и постараемся оптимизировать их для экономии памяти

```python
df_games.info(memory_usage='deep')
```

```python
df_games.head()
```

Самый нежелательный тип данных это `object`, но мы ничего с ним сделать не можем. Однако, мы можем обработать типы данных `float64` и `int64`.<br>
Внутри функции `dtype_optimisation` напишем цикл, который получит названия каждой колонки и проверит ее тип данных. С помощью конструкции `if else` проверим если тип данных равен `int64` или `float64`, и приведем к оптимальному типу данных согласно содержимому колонки методом `to_numeric` и параметром `downcast = 'integer'` и `downcast = 'float'` соответственно.<br>

В случае если возникнет какая нибудь ошибка, вы выведем это на экран благодаря конструкции `try except`.  

```python
def dtype_optimisation(dataframe):
    for column in dataframe.columns:
        if dataframe[column].dtypes == "int64":
            try:
                dataframe[column] = pd.to_numeric(dataframe[column], downcast = 'integer')
            except:
                print(f"Пропущена колонка {column} с типом {dataframe[column].dtypes}")
        elif dataframe[column].dtypes == "float64":
            try:
                dataframe[column] = pd.to_numeric(dataframe[column], downcast = 'float')
            except:
                print(f"Пропущена колонка {column} с типом {dataframe[column].dtypes}")
```

```python
dtype_optimisation(df_games)
```

```python
df_games.info(memory_usage='deep')
```

## Определение портрета пользователя по регионам

- Определим так же и ТОП-5 популярных жанров.
- Узнаем как, и если, влияет рейтинг ESRB на продажи в отдельном регионе.

Для построения графиков, напишем несколько функций для быстрого вывода данных на экран. Каждой функции передадим название региона, а актуальный период будет заложен в теле функции. 

На предыдущем шаге были обнаружены выбросы, отдельные игры, которые сильно влияли на рейтинг игрового жанра. Отсечь такие выбросы получилось ограничив до 15 млн по колонке `total_sales`, добавим это условии и для функций.

`region_top5platforms` - построит график с пятью самыми популярными платформами.

```python
def region_top5platforms(region):
    top5platforms = (
        df_games
        .query('year_of_release >= 2011 and total_sales < 15') #2007
        .groupby('platform', as_index=False)[region].sum()
        .sort_values(by=region, ascending=False)
        .head(5)
    )

    top5platforms.plot(
        kind = 'bar',
        x = 'platform',
        title = 'ТОП-5 платформ',
        figsize = (15, 5),
        grid = True,
        legend = True,
        rot=0,
    )
    plt.ylabel('Количество продаж в млн')
    plt.xlabel('Игровые платформы')
    plt.show()
```

`region_top5genres` - покажет ТОП-5 популярных жанров для этого региона

```python
def region_top5genres(region):
    top5platforms = (
        df_games
        .query('year_of_release >= 2011 and total_sales < 15')
        .groupby('genre', as_index=False)[region].sum()
        .sort_values(by=region, ascending=False)
        .head(5)
    )

    top5platforms.plot(
        kind = 'bar',
        x = 'genre',
        title = 'ТОП-5 жанров',
        figsize = (15, 5),
        grid = True,
        cmap = 'tab20c',
        legend = True,
        rot=0,
    )
    plt.ylabel('Количество продаж в млн')
    plt.xlabel('Игровые жанры')
    plt.show()
```

`region_esrb` - отобразит часто встречающиеся рейтинги за исключением `Unknown`

```python
def region_esrb(region):
    top5platforms = (
        df_games
        .query('year_of_release >= 2011 and rating != "Unknown" and total_sales < 15')
        .groupby('rating', as_index=False)[region].sum()
        .sort_values(by=region, ascending=False)
        .head(5)
    )

    top5platforms.plot(
        kind = 'bar',
        x = 'rating',
        title = 'ТОП-5 рейтингов ESRB',
        figsize = (15, 5),
        grid = True,
        cmap = 'Dark2',
        legend = True,
        rot=0,
    )
    plt.ylabel('Количество продаж в млн')
    plt.xlabel('ESRB рейтинг')
    plt.show()
```

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковы

`Нулевая гипотеза:` Средние пользовательские рейтинги платформ `Xbox One` и `PC` одинаковы.  
`Альтернативная гипотеза:` Средние пользовательские рейтинги платформ `Xbox One` и `PC` **НЕ** одинаковы.

Установим критический уровень статистической значимости равным `alpha = 0.05` согласно общепринятому научному стандарту.  
Если `p-value` окажется меньше него, отвергнем гипотизу.

Т.к. пропуски в `user_score` были заполненны 0, отбросим нулевые рейтинги. Берем данные только за актуальный период.

```python
print(st.ttest_ind(
    df_games.query('platform == "XOne" and user_score > 0 and year_of_release >= 2011')['user_score'], 
    df_games.query('platform == "PC" and user_score > 0 and year_of_release >= 2011')['user_score'], 
    equal_var = False)
.pvalue)
```

Т.к. `p-value` больше установленного критического уровня статистической значимости, мы не можем отвергнуть нулевую гипотезу. Признаем что средние пользовательские рейтинги скорей всего одинаковы.

---

### Средние пользовательские рейтинги жанров Action и Sports разные

`Нулевая гипотеза:` Средние пользовательские рейтинги жанров `Action` и `Sports` одинаковы.  
`Альтернативная гипотеза:` Средние пользовательские рейтинги жанров `Action` и `Sports` **НЕ** одинаковы.

```python
print(st.ttest_ind(
    df_games.query('genre == "Action" and user_score > 0 and year_of_release >= 2011')['user_score'], 
    df_games.query('genre == "Sports" and user_score > 0 and year_of_release >= 2011')['user_score'], 
    equal_var = False)
.pvalue)
```

`p-value` ниже критического уровеня статистической значипости в `0.05` а значит мы можем отвергнуть нулевую гипотезу и средние пользовательские рейтинги для жанров `Action` и `Sports` скорее всего **НЕ** одинаковы.
